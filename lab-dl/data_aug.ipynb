```python
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
import numpy as np
import torch
```

**Visualization**

- [3] Random Affine
- [4] Elastic Transform
- [7] Color Jitter
- [9] Gaussian Blur *(Kernel Size: 7)*

```python
torch.manual_seed(42)

# Define transforms
data_transforms = [
    transforms.RandomAffine(degrees=8, translate=(0.1, 0.1), scale=(0.9, 1.2)),
    transforms.ElasticTransform(alpha=8.0, sigma=0.4),
    transforms.ColorJitter(brightness=0.2, saturation=0.1, hue=0.15),
    transforms.GaussianBlur(kernel_size=7, sigma=(1.0, 2.0))
]

# Convert image to PIL
image_pil = Image.fromarray((image).astype('uint8'))

# List of origin image and augmented images
images = [image_pil]
for transform in data_transforms:
    aug_img = transform(image_pil)
    images.append(aug_img)
```

```python
titles = [
    '[0] Original Image',
    '[3] Random Affine',
    '[4] Elastic Transform',
    '[7] Color Jitter',
    '[9] Gaussian Blur'
]
```

```python
fig, axes = plt.subplots(1, 5, figsize=(15, 3))
for i, ax in enumerate(axes):
    ax.imshow(np.array(images[i]))
    ax.axis('off')
    ax.set_title(titles[i])

plt.show()
```
